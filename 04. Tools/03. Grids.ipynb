This notebook shows how to produce **grids** using **consav.grids**.

## 1. <a id='toc1_'></a>[Setup](#toc0_)

```python
%load_ext autoreload
%autoreload 2

import time
import numpy as np

import matplotlib.pyplot as plt   
colors = plt.rcParams['axes.prop_cycle'].by_key()['color']
plt.rcParams.update({"axes.grid" : True, "grid.color": "black", "grid.alpha":"0.25", "grid.linestyle": "--"})
plt.rcParams.update({'font.size': 14})

# load the module
from consav import grids
```

## 2. <a id='toc2_'></a>[Settings](#toc0_)

```python
ns = [5,10,15,20,25,30]
phis = [1,1.5,2.0]
```

## 3. <a id='toc3_'></a>[`nonlinspace`](#toc0_)

**Illustration:**

```python
for phi in phis:
    
    fig = plt.figure()
    ax = fig.add_subplot(1,1,1)
    ax.set_title(f'phi = {phi}')

    for n in ns:
        x = grids.nonlinspace(0,10,n,phi)
        ax.scatter(n*np.ones(n),x)
```

## 4. <a id='toc4_'></a>[`equilogspace`](#toc0_)

```python
fig = plt.figure()
ax = fig.add_subplot(1,1,1)
ax.set_title(f'equilogspace')

for n in ns:
    x = grids.equilogspace(0,10,n)
    ax.scatter(n*np.ones(n),x)
```
